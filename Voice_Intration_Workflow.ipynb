# ** AI Voice Interaction Flow**

  **Speech-to-Text with Whisper-opensource**  
   - The user speaks, and **Whisper** transcribes the speech into **text (user_text)**.  

  **Text Generation with DeepSeek via Groq-again-opensource**  
   - The transcribed `user_text` is sent to the **DeepSeek model (hosted on Groq)**.  
   - The model processes the input and generates a relevant **AI response (final_response)**.  

  **Text-to-Speech (TTS) with Bark-opensource**  
   - The `final_response` is then passed to the **Bark model**.  
   - Bark converts the AI-generated text back into **speech**, allowing the user to hear the response.

```python
import whisper  #load the whisper which is used for automatic speech recognition speech to text.

model = whisper.load_model("medium")

def transcribe(audio_path):
  text = model.transcribe(audio_path)
  return text

result = transcribe("/content/bark_generation.wav")
```

```python
from groq import Groq
import os

os.environ["GROQ_API_KEY"] = userdata.get('groq_api_key')         #get the groq api here.

client = Groq()
completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages = [
    {"role": "system", "content": "You are a helpful assistant. Provide specific and to-the-point responses. If you don't have an answer, say 'Sorry, I can't assist you with this question. You may ask another question.'"},
    {"role": "user", "content": result["text"]}
    ],
    temperature=1,
    max_completion_tokens=1024,
    top_p=1,
    stream=True,
    stop=None,
)
```

```python
print(completion)
```

```python
def getChatResponse(completion):  #get the data from the chnks.
    final_text = []
    for chunks in completion:
        final_text.append(chunks.choices[0].delta.content or "")
    return "".join(final_text)
```

```python
final_response=getChatResponse(completion)
print(final_response)
```

```python
from bark import SAMPLE_RATE, generate_audio, preload_models  #bark will work when model will gave back reponse in the form of final_text it will take the final reponse and covert them into speech again.
from scipy.io.wavfile import write as write_wav
from IPython.display import Audio

# download and load all models
preload_models()

# generate audio from text
text_prompt = f"{final_response}"
audio_array = generate_audio(text_prompt)

write_wav("Bark-response", SAMPLE_RATE, audio_array)

Audio(audio_array, rate=SAMPLE_RATE)
```
